```python
from scipy.stats import multivariate_normal
def gauss_fun(X, Y):
    mux = [0]
    muy = [0]
    sx = [1]
    sy = [3]
    rho = [0]
    d = np.dstack([X,Y])
    z = None
    for i in range(len(mux)):
        mean = [mux[i],muy[i]]
        # Extract covariance matrix
        cov = [[sx[i] * sx[i], rho[i] * sx[i] *sy[i]], [rho[i] * sx[i] * sy[i], sy[i] * sy[i]]]
        gaussian = multivariate_normal(mean=mean,cov=cov)
        z_ret = gaussian.pdf(d)
        if z is None:
            z = z_ret
        else:
            z +=z_ret
    return z

x = np.linspace(-7, 7, 100)
y = np.linspace(-7, 7, 100)
X, Y = np.meshgrid(x, y)
Z = gauss_fun(X, Y)

plt.contour(X, Y, Z, colors='black');
```

a) Ratio is 3.

b) Covariance increases. Largest variance decreases while smallest goes up.

# Exercise 1.15

a) The optimal decision boundary has linear shape.

```python
myData = pr.gendats((10,10), dim=2, delta=1)
labelData = pr.labeld(myData)
# estimate linear discriminants by means of ldc(a) 
# and quadratic discriminant boundaries by means of qdc(a).
myLDC = pr.ldc()
myQDC = pr.qdc()
myLDC.train(myData)
myQDC.train(myData)

pr.scatterd(myData)
pr.plotc(myLDC)
pr.plotc(myQDC)
```

b) \
The boundary shape will be quadratic because of the limited training points.

c) \
LDC is by construction linear, while QDC is nonlinear as the decision boundary.

d) \
When the number of points per class increases, the solutions by ldc and qdc will coincide.\
When there are two samples for each class, the number of the samples is not enough for QDC estimator.\
When there is only one sample for each class, it is not enough for LDC estimator.\
Therefore, the quantity of estimating samples for QDC and LDC is more than 2 and more than 1 respectively.

```python
set1 = pr.numpy.random.randn(10,2)
label1 = np.zeros((1,10))
set2 = pr.numpy.random.randn(10,2)
label2 = np.ones((1,10))

setAll = np.vstack([set1,set2])
labelAll = np.vstack([label1.T,label2.T])

myData = pr.prdataset(setAll, labelAll)

myLDC = pr.ldc()
myQDC = pr.qdc()

myLDC.train(myData)
myQDC.train(myData)

pr.scatterd(myData)
pr.plotc(myLDC)
pr.plotc(myQDC)

errorLDC = pr.testc(myLDC(myData))
errorQDC = pr.testc(myQDC(myData))

# points on the wrong side of the boundary
print("errorLDC:", errorLDC*20)
print("errorQDC:", errorQDC*20)
```

```python
# Generate a simple 2D dataset (for instance, using gendatb) and 
# plot the decision boundaries of the six classiers listed above. 
# Use k = 1 for the knnc.
dat = pr.prdataset(pr.gendatb(n=(50,50),s=2))
NMC = pr.nmc(dat)
LDC = pr.ldc(dat)
QDC = pr.qdc(dat)
FisherC = pr.fisherc(dat)
ParzenC = pr.parzenc(dat)
KNN = pr.knnc(dat,(1))

pr.scatterd(dat)
pr.plotc(NMC)
pr.plotc(LDC)
pr.plotc(QDC)
```

```python
pr.scatterd(dat)
pr.plotc(FisherC)
pr.plotc(ParzenC)
pr.plotc(KNN)

for C in [NMC,LDC,QDC,FisherC,ParzenC,KNN]:
    print(str(C),":",pr.testc(C(dat)))
    print()
```

```python
# Make a new dataset in which the second feature is 10 times larger than the dataset above. 
newtrain = dat;
newtrain[:,1] = 10*(+newtrain[:,1]).reshape(100)

NMC = pr.nmc(newtrain)
LDC = pr.ldc(newtrain)
QDC = pr.qdc(newtrain)
FisherC = pr.fisherc(newtrain)
ParzenC = pr.parzenc(newtrain)
KNN = pr.knnc(dat,(1))
```

```python
pr.scatterd(newtrain)
pr.plotc(NMC)
pr.plotc(LDC)
pr.plotc(QDC)
```

```python
pr.scatterd(newtrain)
pr.plotc(FisherC)
pr.plotc(ParzenC)
pr.plotc(KNN)
```

```python
for C in [NMC,LDC,QDC,FisherC,ParzenC,KNN]:
    print(str(C),":",pr.testc(C(newtrain)))
    print()
```

c) The QDC, LDC and Fisher discriminant are not affected by this rescaling of the feature space, while the NMC and the k-NN are affected by it. It is because in the first classifiers a covariance matrix is estimated, thus estimating the scale of the feature values automatically in these classifiers.

d) In many cases it is an advantage, because it means that you don't need to optimize the scaling of the features. The disadvantage is that the trainig set should be large enough to make the estimation of the scale reliable.
